### What is a Hidden Markov Model (HMM)?

A Hidden Markov Model (HMM) is a probabilistic model used to represent patterns that evolve over time or position â€” such as DNA, RNA, or protein sequences.  
In bioinformatics, HMMs are powerful tools for modeling sequence families and detecting homology between proteins.

---

#### ðŸ”¹ Core Idea
An HMM does not align sequences directly.  
Instead, it learns the statistical properties of a family of related sequences by modeling:
- Match states â€“ conserved amino acids that appear frequently at certain positions,  
- Insert states â€“ positions where extra residues may occur,  
- Delete states â€“ positions that can be skipped.

Each state is associated with probabilities of emitting particular amino acids or transitioning to another state.

---

#### ðŸ”¹ What HMM Does in Sequence Analysis
1. Training (Building the Model):
   - From a Multiple Sequence Alignment (MSA), HMM learns which residues are conserved or variable.
   - It captures the probability of seeing each amino acid at each position.

2. Searching (HMMER / hmmsearch):
   - The trained HMM is compared to new sequences.
   - The algorithm computes a score or E-value that indicates how likely it is that the new sequence belongs to the same family.

3. Interpretation:
   - Low E-value â†’ high similarity â†’ probably from the same family.  
   - High E-value â†’ low similarity â†’ likely unrelated.

---

#### ðŸ”¹ Why Itâ€™s Useful
- Detects distant evolutionary relationships that simple alignments may miss.  
- Handles insertions, deletions, and variable lengths gracefully.  
- Generates compact, reusable models to compare new proteins against.

---

> ðŸ§  In short:  
> HMM transforms a set of aligned sequences into a *statistical fingerprint* of the family â€”  
> allowing us to recognize new members even if they are only weakly similar at the sequence level.

## ðŸ§¬   Multiple Sequence Alignment (MSA)

A **Multiple Sequence Alignment (MSA)** is created for each group of homologous sequences.  
MSA helps reveal conserved regions and variable sites that are important for family modeling.

These aligned sequences form the basis for building the **HMM profiles**.

---


## âš™ï¸  Building the HMM Profile

Each MSA is converted into a **Hidden Markov Model profile**.  
The HMM represents:
- The probability of each amino acid at each position  
- The likelihood of insertions or deletions  
- The conserved and variable regions of the family  

This model will later help measure how well new sequences fit into the same family.

## ðŸ§® Mathematical Background of HMM Scoring

When building an HMM profile from aligned sequences, the model can be evaluated using three key statistical scores: MSV, Viterbi, and Forward.  
Each of these scores reflects a different way of measuring how well a new sequence fits the trained model.

---

### 1ï¸âƒ£ MSV (Multiple Segment Viterbi)

- Purpose: A fast pre-filter that estimates the similarity between a sequence and the model without considering insertions or deletions.  

- Calculation:

$$
S_{MSV} = \sum_{i=1}^{L} \log \frac{P(x_i \mid M_i)}{P(x_i \mid \text{background})}
$$

where:  

- \( P(x_i \mid M_i) \): emission probability of amino acid \(x_i\) in match state \(M_i\)  
- \( P(x_i \mid \text{background}) \): probability of observing the same residue randomly  

- Interpretation:  
  A higher (less negative) MSV score indicates stronger similarity between the sequence and the model.

---

### 2ï¸âƒ£ Viterbi Score

- Purpose: Finds the *most probable alignment path* (including matches, insertions, and deletions).  

- Recursive formula:

$$
V_j(i) = \max \begin{cases}
V_{j-1}(i-1) + \log a_{MM} + \log e_M(x_i) \\
V_{j}(i-1) + \log a_{IM} + \log e_M(x_i) \\
V_{j-1}(i) + \log a_{DM} + \log e_M(x_i)
\end{cases}
$$

where:  
- \( a_{XY} \): transition probability from state \(X\) to \(Y\)  
- \( e_M(x_i) \): emission probability of residue \(x_i\) in match state  

- Interpretation:  
  Represents the log-probability of the *best* alignment path between the sequence and the HMM.

---

### 3ï¸âƒ£ Forward Score

- Purpose: Calculates the *total probability* of the sequence by summing over *all possible alignment paths*.  

- Recursive formula:

$$
F_j(i) = \left[ F_{j-1}(i-1)a_{MM} + F_{j}(i-1)a_{IM} + F_{j-1}(i)a_{DM} \right] e_M(x_i)
$$

- Interpretation:  
  Unlike Viterbi, Forward sums probabilities instead of taking the maximum.  
  Therefore, its value is always higher (less negative) as it includes all possible alignments.

---

### ðŸ§© Summary Table

| Metric | Operation | Considered Paths | Description |
|---------|------------|------------------|--------------|
| MSV | Summation of log-odds | Match states only | Fast, approximate similarity |
| Viterbi | Maximum probability | Single best path | Most probable alignment |
| Forward | Summation over all probabilities | All possible paths | Overall likelihood of similarity |

---

In our HMM output:  
- MSV provides a quick estimate of similarity,  
- Viterbi refines the alignment,  
- Forward measures total probability that the sequence belongs to the same family.

### ðŸ§  Interpretation of HMM Output for Hemoglobin Alpha Sequences

After generating a Hidden Markov Model (HMM) from the MSA of the three hemoglobin alpha sequences (Horse, Cow, and Pig),  
the output provides a position-specific statistical representation of the alignment.

---

#### ðŸ” What the HMM Output Represents

1. Match States (M1, M2, â€¦)  
   Each column of the MSA that contains mostly amino acids (not gaps) becomes a â€œmatch state.â€  
   These states represent conserved positions â€” residues that are common across all sequences (e.g., key histidines that bind heme).  

2. Insertion States (I1, I2, â€¦)  
   Columns with extra residues in some sequences (gaps in others) are modeled as â€œinsertion states.â€  
   They allow flexibility â€” small evolutionary variations in sequence length.

3. Transition Probabilities  
   The HMM encodes probabilities of moving between states (e.g., Match â†’ Match, Match â†’ Insert, Insert â†’ Match).  
   High transition probability from M to M indicates strong conservation; high M â†’ I means structural variability.

4. Emission Probabilities  
   Each match state includes emission probabilities for the 20 amino acids.  
   A high emission probability for a residue (e.g., Valine at position 10) means itâ€™s conserved across species.

---

#### ðŸ“Š Overall Interpretation

- The model shows strong conservation across most positions â€” as expected for hemoglobin alpha chains,  
  which are evolutionarily stable and differ only slightly between mammals.  
- Variable regions correspond to surface loops or non-critical residues, allowing substitutions (e.g., Gly â†” Ala).  
- The final section of the HMM (near the C-terminus) may show lower conservation, reflecting relaxed structural constraints.

---

ðŸ§© In summary, the HMM captures both the conserved â€œcore patternâ€ of the hemoglobin alpha family  
and the permissible variability, making it more robust than raw sequence alignment methods like Clustal.

### Comparison: Clustal Omega (MSA) vs. HMM Profile

When analyzing the Horse, Cow, and Pig Hemoglobin Alpha sequences,  
both Clustal Omega and Hidden Markov Model (HMM) methods were applied,  
but they serve different purposes and levels of generalization.

ðŸ”¹ Clustal Omega (MSA)
- Performs a *direct multiple sequence alignment* of the input sequences.  
- Shows exact residue positions, gaps (`-`), and conserved regions (`*`, :, `.`).  
- Provides a visual and pairwise comparison between the given sequences.  
- Focuses only on the sequences provided â€” it cannot generalize to new ones.

ðŸ”¹ HMM (Hidden Markov Model)
- Builds a statistical profile based on the aligned sequences.  
- Learns the probability of each residue and the likelihood of insertions/deletions.  
- Can be used to search for or classify new sequences from the same family.  
- Abstracts away from the raw alignment and captures evolutionary patterns.

ðŸ”¹ Summary
| Aspect | Clustal Omega | HMM |
|--------|----------------|-----|
| Type | Alignment tool | Probabilistic model |
| Input | Raw sequences | MSA result |
| Output | Alignment visualization | Profile model |
| Purpose | Compare given sequences | Detect family membership |
| Generalization | No | Yes |
| Example use | Show similarity between Horse, Cow, Pig | Test if Sheep sequence belongs to same family |

> In short: Clustal Omega shows *where* the sequences align,  
> while HMM captures *how* the family behaves statistically and can recognize new members.

### Using Hemoglobin Alpha Sequences (Human, Gorilla, and Orangutan)

After aligning three hemoglobin alpha sequences â€” *Human, Gorilla,* and *Orangutan* â€” using Clustal Omega,  
we generated an HMM profile from the same multiple sequence alignment (MSA).  
Below is a comparison of how both methods interpret evolutionary conservation and sequence similarity.

---

### ðŸ”¹ Clustal Omega (MSA Alignment)

Concept:  
Clustal Omega aligns sequences directly by comparing residues position by position, using pairwise and progressive alignment.  

What it shows:  
A clear, *visual alignment* where:
- * marks identical residues,  
- : indicates conservative substitutions,  
- . represents semi-conservative changes.  

Strengths:  
- Fast, intuitive, and visually interpretable.  
- Ideal for comparing highly similar sequences.  

Limitations:  
- Captures alignment but not residue-level probabilities.  
- Less effective at detecting distant homologs.  

ðŸ§  In our hemoglobin dataset:  
Clustal Omega produced an *almost perfect alignment* â€” every residue aligned identically across the three primate species.  
This indicates strong evolutionary conservation, confirming that the alpha-globin chain has remained virtually unchanged among primates.

---

### ðŸ”¹ Hidden Markov Model (HMM Profile)

Concept:  
The HMM converts the MSA into a *statistical model* that describes how likely each amino acid (and gaps) is at every position.  
It captures both conservation and variation in a probabilistic way.

What it shows:  
- Match (M) states: Highly conserved residues across all sequences.  
- Insert (I) and Delete (D) states: Rare variations or gaps.  
- Transition probabilities: Likelihood of moving between states (Mâ†’M, Mâ†’I, etc.).  

Strengths:  
- Detects remote homologs through statistical modeling.  
- Captures subtle evolutionary information beyond visual similarity.  

Limitations:  
- Output is numeric and less intuitive to read.  
- Requires a clean, high-quality MSA as input.  

ðŸ§  In our hemoglobin model:  
The HMM shows very high match probabilities and almost zero insertions or deletions,  
indicating that these sequences form a high-confidence conserved domain.  
If a new hemoglobin-like sequence were tested, it would receive a very high alignment score,  
whereas unrelated proteins (e.g., myoglobin or insulin) would yield significantly lower scores.

---

## ðŸ§¬ BLAST vs PSI-BLAST

BLAST (Basic Local Alignment Search Tool)  
Compares a query sequence against a database to find regions of local similarity.  
- Used to identify close homologs or annotate unknown sequences.  
- Fast and accurate for detecting high-identity matches.  
- Example: blastp compares a protein against a protein database.

PSI-BLAST (Position-Specific Iterated BLAST)  
An advanced version of BLAST that increases sensitivity through iterations.  
- Builds a profile (PSSM) from the first BLAST hits.  
- Re-searches the database using this profile to detect remote homologs that normal BLAST might miss.  
- Useful for finding proteins from the same family with low sequence identity.

Main differences:  
| Feature | BLAST | PSI-BLAST |
|----------|--------|-----------|
| Type | Single search | Iterative search (2â€“5 rounds) |
| Sensitivity | Detects close homologs | Detects distant homologs |
| Output | Alignments, scores, E-values | Same + profile (PSSM) |
| Speed | Faster | Slower (due to iterations) |

Summary:  
- Use BLAST for quick searches and clear matches.  
- Use PSI-BLAST when you need to uncover distant evolutionary relationships.
