<h3>PDB<h3>

PDB (Protein Data Bank) : 
It’s the main worldwide repository for 3D structural data of biological macromolecules (proteins, DNA, RNA, and complexes).<br>
It’s like a “library” where scientists deposit experimentally determined structures from techniques like:<br>
X-ray crystallography (most entries),

NMR spectroscopy,

Cryo-electron microscopy (cryo-EM).

<h3>Extract protein sequences (FASTA) from chains<h3>

Polypeptide (PP) builder — it scans chains and constructs continuous polypeptide sequences from coordinates (returns Polypeptide objects).

```python
from Bio.PDB import PPBuilder
```

```python
ppb = PPBuilder()
fasta_out = "chains.fasta"
with open(fasta_out, "w") as fh:
    for model in structure:
        for chain in model:
            peptides = ppb.build_peptides(chain)
            if not peptides:
                continue
            seqs = [str(p.get_sequence()) for p in peptides]
            seq = "".join(seqs) if len(seqs) > 1 else seqs[0]
            header = f">_{chain} model{model.id} len={len(seq)}"
            fh.write(header + "\n")
            fh.write(seq + "\n")
```

<h2>working with RNA families <h2>

```python
from Bio import AlignIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import urllib.request
```

```python
rfam_id = "RF00027"  
url = f"https://rfam.org/family/{rfam_id}/alignment?format=stockholm"
stockholm_file = f"{rfam_id}.sto"
urllib.request.urlretrieve(url, stockholm_file)
```

```python

alignment = AlignIO.read(stockholm_file, "stockholm")
print(f"Number of sequences in alignment: {len(alignment)}")
print(f"Alignment length: {alignment.get_alignment_length()}")

seq_records = []
for record in alignment:
    ungapped_seq = str(record.seq).replace("-", "")
    seq_records.append(SeqRecord(Seq(ungapped_seq),
                                 id=record.id,
                                 description="Rfam tRNA sequence"))

fasta_file = f"{rfam_id}.fasta"
SeqIO.write(seq_records, fasta_file, "fasta")
print(f"FASTA file saved as {fasta_file}")
```
